```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_excel("kidney_disease.xlsx")

pd.set_option('display.max_columns',None)

df
```

```python
#Preprocessing

df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dropna(how='all')
```

```python
df = df[~df['pcv'].isin(['?'])]
df.pcv.unique()
```

```python
df.drop(['id'], axis=1, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.select_dtypes(include=['float64', 'int64'])
```

```python
df.select_dtypes(include=['object'])
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
df['wc'] = df['wc'].astype('float64')
df['rc'] = df['rc'].astype('float64')
```

```python
df.info()
```

```python
df
```

```python
# Select only numeric columns for histogram plotting
df.select_dtypes(include=['float64', 'int64']).hist(figsize=(12, 8), bins=30)
plt.tight_layout()
plt.show()
```

```python
#Numerical Features
df['age'] = df['age'].fillna(df['age'].mean())
df['bp'] = df['bp'].fillna(df['bp'].median())
df['sg'] = df['sg'].fillna(df['sg'].mean())
df['al'] = df['al'].fillna(df['al'].median())
df['su'] = df['su'].fillna(df['su'].median())
df['bgr'] = df['bgr'].fillna(df['bgr'].median())
df['bu'] = df['bu'].fillna(df['bu'].median())
df['sc'] = df['sc'].fillna(df['sc'].median())
df['sod'] = df['sod'].fillna(df['sod'].median())
df['pot'] = df['pot'].fillna(df['pot'].median())
df['hemo'] = df['hemo'].fillna(df['hemo'].mean())
df['pcv'] = df['pcv'].fillna(df['pcv'].mean())
df['wc'] = df['wc'].fillna(df['wc'].median())
df['rc'] = df['rc'].fillna(df['rc'].mean())

# Categorical Features
df['rbc'] = df['rbc'].fillna(df['rbc'].mode()[0])
df['pc'] = df['pc'].fillna(df['pc'].mode()[0])
df['pcc'] = df['pcc'].fillna(df['pcc'].mode()[0])
df['htn'] = df['htn'].fillna(df['htn'].mode()[0])
df['dm'] = df['dm'].fillna(df['dm'].mode()[0])
df['cad'] = df['cad'].fillna(df['cad'].mode()[0])
df['appet'] = df['appet'].fillna(df['appet'].mode()[0])
df['pe'] = df['pe'].fillna(df['pe'].mode()[0])
df['ane'] = df['ane'].fillna(df['ane'].mode()[0])
df['ba'] = df['ba'].fillna(df['ba'].mode()[0])


```

```python
df.isnull().sum()
```

```python
df.rename(columns={'classification':'kidney_disease'},inplace=True)
```

```python
df
```

```python
# Encoding Categorical Features
from sklearn.preprocessing import LabelEncoder


catrgorical_column = df.select_dtypes(include='object').columns

catrgorical_column
```

```python
encoders={}

for col in catrgorical_column:
    encoders[col] = LabelEncoder()
    df[col] = encoders[col].fit_transform(df[col])
    
```

```python
encoders
```

```python
#Feature selection

df.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',vmin=-1,vmax=1)
plt.show()
```

```python
correaltion_matrix = df.corr()

thershold =0.3

target_column = "kidney_disease"

high_corr = correaltion_matrix[target_column][correaltion_matrix[target_column].abs()>thershold].index.tolist()

high_corr.remove(target_column)

print(high_corr)
```

```python
df_feature_selection = df[['sg', 'al', 'pc', 'bgr', 'bu', 'sod', 'hemo', 'pcv', 'rc', 'htn', 'dm', 'appet', 'pe', 'ane']]
df_feature_selection
```

```python
# Splitting feature(x) and Target(y)

x = df_feature_selection
y= df['kidney_disease']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Model fitting

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train,y_train)
```

```python
# Model Evaluation

lr.score(x_test,y_test) #r2 / slope
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

con = confusion_matrix(y_test,y_pred=lr.predict(x_test))
con
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(con,annot=True,cmap='coolwarm')
plt.xlabel('predicted')
plt.ylabel('actual')
plt.show()
```

```python
class_report = classification_report(y_test,y_pred=lr.predict(x_test))
print(class_report)
```

```python
accuracy = accuracy_score(y_test,y_pred=lr.predict(x_test))
accuracy
```
